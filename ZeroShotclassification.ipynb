```python
import pandas as pd

```

```python
from transformers import pipeline
```

instantiate pipeline

```python
zsc_pipeline=pipeline("zero-shot-classification",model="facebook/bart-large-mnli")

zsc_pipeline
```

```python
zsc_pipeline
```

```python
text ="This thrilling novel follows a detective as he unravels a complex murder mystery in a small town."

print(text)

results=zsc_pipeline(text, candidate_labels=['mystery','crime','suspense','romance','comedy'])
results
```

```python
df = pd.DataFrame({'labels':results['labels'], 'scores':results['scores']})
df
```

```python
text ="As the spaceship approached the alien planet, the crew prepared for an encounter with an unknown extraterrestrial species."

print(text)

results=zsc_pipeline(text, candidate_labels=['education','space exploration','programming'])
results

df = pd.DataFrame({'labels':results['labels'], 'scores':results['scores']})
df
```
